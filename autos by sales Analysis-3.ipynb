```python
#  3) What is the Average power of a vehicle by vehicle type and gearbox type.Give a plot
```

```python
avg_power=datac.groupby(["vehicleType","gearbox"])["powerPS"].mean()
avg_power.plot(kind="bar")
plt.xlabel("vehicleType and gearbox type")
plt.ylabel("average power")
plt.title("avg power of vehicletype and gearbox")
plt.show()
```
